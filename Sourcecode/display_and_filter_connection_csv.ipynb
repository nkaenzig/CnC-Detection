# Display and Filter connections.csv using Pandas dataframes 

```python
import pandas as pd
import os
import numpy as np
# import qgrid

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))


pd.set_option("display.max_rows", None)

extracted_path = './extracted/'
connections_csv = 'connections.csv'

df = pd.read_csv(os.path.join(extracted_path, connections_csv), sep=';')
# qgrid_widget = qgrid.show_grid(df, show_toolbar=True)
# qgrid_widget
```

```python
# qgrid_widget.get_changed_df()
```

### Drop Columns

```python
# Drop columns
df = df.drop('min_duration', axis=1)
df = df.drop('max_duration', axis=1)
```

### Sorting

```python
df = df.sort_values(by='nr_connections', ascending=True)
```

### Filtering & displaying 
Note: str.contains() accepts regular expressions, e.g. 'external|internal'

```python
# %%time
# df_f = df[df['src_ip']=='10.7.2.2']

filter = df['src_ip'].str.contains('babe')
filter = filter & df['dst_ip'].str.contains('')
filter = filter & df['src_class'].str.contains('external|unknown')
filter = filter & df['dst_class'].str.contains('internal')
filter = filter & df['conn_states'].str.contains('S0|REJ')

df_f = df[filter]
df_f

```

### Print unique column entries

```python
for element in df_f['dst_ports'].unique():
    print(element)
```
